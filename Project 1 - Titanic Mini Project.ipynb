```python
# Separating Child, Male and Female:

def male_female_child(passenger):
   # print("Passenger:",passenger)
    Age,Sex = passenger
    if (Age < 18):
        return("Child")
    else:
       return Sex
```

```python
titanic_df['Sex_2'] = titanic_df[['Age','Sex']].apply(male_female_child, axis =1 )
```

```python
titanic_df['Sex'].isna().head(7)
```

```python
titanic_df['Sex_2'] = titanic_df['Sex']

titanic_df.head(10)
```

```python
titanic_df.Sex_2.where((titanic_df['Age'].isna() == True) | (titanic_df.Age > 18) ,"Child", inplace = True)
```
